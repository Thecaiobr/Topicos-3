```python
from sklearn.datasets import fetch_20newsgroups

# Carregar o conjunto de dados
newsgroups_data = fetch_20newsgroups(subset='train')
```

```python
import re

def get_tokens(text):
    tokens = re.sub("[^\w]", " ", text).split()
    cleaned_text = [w.lower() for w in tokens]
    return cleaned_text

def tokenize(texts):
    words = []
    for text in texts:
        w = get_tokens(text)
        words.extend(w)
    words = sorted(list(set(words)))
    return words

def preProcessCorpus(corpus):
  newCorpus = [doc.lower() for doc in corpus]
  regex = r"(?<!\d)[\!?.,;:-](?!\d)"
  newCorpus = [re.sub(regex, "", doc, 0) for doc in newCorpus]
  return newCorpus

# Extrair os textos do conjunto de dados
# ESSES VALORES PODEM SER ALTERADOS entre 0 e 20 sendo [0:20] para rodar todos textos)
# No codigo abaixo esta contido o codigo para rodar os 3 primeiros conjuntos de texto
C = newsgroups_data.data[:3]

print("C has %d texts:" % len(C))
for i in range(len(C)):
    print('t{} = {}'.format(i+1, C[i]))

# Tokenizar os textos para obter o vocabulário
V = tokenize(C)
print("V has {0} words: {1}".format(len(V), V))

#ONE-HOT ENCODING
import numpy as np

with open("20News_01.txt", "w") as file:
  for text in C:
      words = get_tokens(text)
      bag_vector = np.zeros(len(V))
      for w in words:
          for i, word in enumerate(V):
              if word == w:
                  bag_vector[i] = 1
      print("{0} = {1}".format(text, np.array(bag_vector)))
      file.write("{0} = {1}\n".format(text, np.array(bag_vector)))
```

```python
#COUNT VECTORS
with open("20News_02.txt", "w") as file:
  for text in C:
      words = get_tokens(text)
      bag_vector = np.zeros(len(V))
      for w in words:
          for i, word in enumerate(V):
              if word == w:
                  bag_vector[i] += 1
      print("{0} = {1}".format(text, np.array(bag_vector)))
      file.write("{0} = {1}\n".format(text, np.array(bag_vector)))
```

```python
#TF-IDF Vectors
import sklearn
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer

corpus = preProcessCorpus(C)
print(corpus)

vectorizer = CountVectorizer()
docTermMatrix = vectorizer.fit_transform(corpus)
terms = vectorizer.get_feature_names_out()

print(pd.DataFrame(docTermMatrix.A, columns=terms).to_string())
```

```python
#TF-IDF Vectors
from sklearn.feature_extraction.text import TfidfTransformer

transformer = TfidfTransformer()
tfIdMatrix = transformer.fit_transform(docTermMatrix)

print(pd.DataFrame(tfIdMatrix.A, columns=terms).to_string())

with open("20News_03.txt", "w") as file:
  file.write(pd.DataFrame(tfIdMatrix.A, columns=terms).to_string());
```

```python
#CO-OCCURRENCE VECTORS
co_occurrence_matrix = (docTermMatrix.T * docTermMatrix)
print(co_occurrence_matrix.todense())
```

```python
import scipy.sparse as sp
g = sp.diags(1./co_occurrence_matrix.diagonal())
co_occurrence_matrix_norm = g * co_occurrence_matrix

print(co_occurrence_matrix_norm.todense())

with open("20News_04.txt", "w") as file:
    file.write("Co-Occurrence Matrix:\n")
    file.write("{0}\n\n".format(co_occurrence_matrix.todense()))

    file.write("Normalized Co-Occurrence Matrix:\n")
    file.write("{0}\n".format(co_occurrence_matrix_norm.todense()))
```

```python
#N-GRAMS
vectorizer = CountVectorizer(ngram_range=(2,2))
docTermMatrix = vectorizer.fit_transform(corpus)
vocabulary = vectorizer.get_feature_names_out()
terms = vectorizer.get_feature_names_out()

print(pd.DataFrame(docTermMatrix.A, columns=terms).to_string())

with open("20News_05.txt", "w") as file:
    file.write(pd.DataFrame(docTermMatrix.A, columns=vocabulary).to_string())
```
